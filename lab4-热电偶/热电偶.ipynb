# 热电偶实验
本程序仅包含“测量热电偶系数”和“验证中间导体定律”的内容，“验证中间温度定律”的内容未包含在内。

## 数据记录
U1 为无中间导体情况下，冷端温度为 0 摄氏度，热端温度为 30-70 摄氏度时每 5 摄氏度记录的热电势值。

U2 为有中间导体情况下，冷端温度为 0 摄氏度，热端温度为 30-70 摄氏度时每 5 摄氏度记录的热电势值。

```python
# data record
import numpy as np
T = np.arange(30, 71, 5)
U1 = np.array([1.95, 2.27, 2.59, 2.91, 3.22, 3.54, 3.86, 4.18, 4.50])
U2 = np.array([1.95, 2.26, 2.58, 2.90, 3.22, 3.54, 3.86, 4.19, 4.51])
```

```python
# use least square method to fit the data
C1 = np.polynomial.polynomial.polyfit(T, U1, 1)
C2 = np.polynomial.polynomial.polyfit(T, U2, 1)

print("C1 = ", C1[1])
print("C2 = ", C2[1])
```

```python
# plot the U-T graph, display the fitting curve, the data points and the fitting parameters
import matplotlib.pyplot as plt
plt.plot(T, U1, 'ro', label='U1')
plt.plot(T, U2, 'bo', label='U2')
plt.plot(T, np.polynomial.polynomial.polyval(T, C1), 'r-', label='U1 fitting')
plt.plot(T, np.polynomial.polynomial.polyval(T, C2), 'b-', label='U2 fitting')
plt.legend()
plt.xlabel('T (degrees Celsius)')
plt.ylabel('U (mV)')
plt.title('U-T graph')

# display the k and b
plt.text(40, 3.7, 'C1 = %.4f' % C1[1], fontdict={'size': 10, 'color': 'red'})
plt.text(40, 3.5, 'C2 = %.4f' % C2[1], fontdict={'size': 10, 'color': 'blue'})

# set dpi to 600
plt.savefig('U-T.png', dpi=600)
plt.show()
```
